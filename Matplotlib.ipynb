Customizing of line plot

```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

How to craete bar plot ? (Vertical)

```python
#importing library
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#creating dataset
data={'apples':10,'oranges':15,'lemons':5,'limes':20}
names=list(data.keys())
values=list(data.values())
#add figure
fig=plt.figure(figsize=(10,5))
#sub-plotting
ax1=plt.subplot()
#plot
ax1.bar(names,values)
#custmization title
plt.title("Vertical Bar Plot")
#cust. the x-axis label and y-axis label
plt.xlabel("Names")
plt.ylabel("Values")
#showing plot
plt.show
```

How to craete bar plot ? (Horizantal)

```python
#creating dataset
data={'apples':10,'oranges':15,'lemons':5,'limes':20}
names=list(data.keys())
values=list(data.values())
#add figure
fig=plt.figure(figsize=(10,5))
#sub-plotting
ax1=plt.subplot()
#plot
#ax1.bar(names,values)
#custmize-hori
ax1.barh(names,values,color='orange')
#custmization title
plt.title("Horizantol Bar Plot")
#cust. the x-axis label and y-axis label
plt.xlabel("Quantity")
plt.ylabel("Fruits")
#cutmi.add grid
#plt.grid(True)
#saving plot
plt.savefig('HorizantalBarPlot.png')
#showing plot
plt.show
```

How to create a scatter plot?

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#creating dataset
a=[10,20,30,40,50,60,70,80]
b=[5,2,4,2,1,4,5,2]
x=[1,2,3,4,3,1,2,2]
#creating a scatter plot
fig=plt.figure(figsize=(10,5))
ax1=plt.subplot()
ax1.scatter(a,b)
ax1.scatter(a,x)
plt.title("Scatter Plot")
plt.legend(['b','x'],loc='best')
#plt.grid(True)
plt.show()
```

Customizing the Scatter plot:

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#creating dataset
a=[10,20,30,40,50,60,70,80]
b=[5,3,2,5,6,1,4,2]
x=[1,2,3,4,3,2,1,3]
fig=plt.figure(figsize=(10,5))
ax1=plt.subplot()
ax1.scatter(a,b, c='g',s=200,marker='o',alpha=0.5)
ax1.scatter(a,x, c='r',s=400,marker='4',alpha=1)
plt.legend(['b','x'], loc='best')
plt.title("Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show()
```

```python
ax1=plt.subplot()
ax1.scatter(a,b, c='g',s=100,marker='o',alpha=0.5)
ax1.scatter(a,x, c='r',s=50,marker='D',alpha=1)
plt.legend(['b','x'], loc='best')
plt.title("Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.savefig('scatter plot.png')
plt.show()
```

How to create a box plot?

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#creating dataset
total=[20,4,1,30,20,10,20,70,30,10]
order=[10,3,1,15,17,2,30,44,2,1]
discount=[30,10,20,5,10,20,50,60,20,45]
data=list([total,order,discount])
#plotting the data
plt.boxplot(data, showmeans =True)
plt.title("Box Plot")
plt.grid(True)
plt.show()
```

How to create a Quiver plot?

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#prepare data
x_pos=[0,0]
y_pos=[0,0]
x_direct=[1,0]
y_direct=[1,-1]
fig=plt.figure(figsize=(10,5))
ax1=plt.subplot()
#plot
ax1.quiver(x_pos,y_pos,x_direct,y_direct,scale=5)
ax1.axis([-1.5,1.5,-1.5,1.5])
plt.title("Quiver plot")
plt.grid(True)
plt.show()
```
